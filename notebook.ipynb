```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analyze_correlation(x, y):
    # This gets the (1 + 1 = 2) coefficients for the line of best fit
    trend = np.polyfit(x, y, 1)
    # print(trend)
    # This evaluates the line of best fit at each x value
    trendline = np.polyval(trend, x)

    # This function accepts many arrays
    # It returns a matrix of the correlation coefficients between each pair of arrays
    # The matrix is symmetric (Hermitian in general), and the diagonal elements are 1
    # Here the matrix is 2x2, because we are passing in 2 arrays
    # We want the off diagonal element, which is the correlation between the two arrays
    correlation = np.corrcoef([x, y])[0, 1]
    print(correlation)

    plt.scatter(x, y)
    plt.plot(x, trendline, color='green')
    plt.show()
```

```python
# Perfect positive correlation
# p=1
x = [1, 2, 3, 4, 5]
y = [20, 40, 60, 80, 100]
analyze_correlation(x, y)
```
